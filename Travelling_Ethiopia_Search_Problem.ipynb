1.2 Write a class that takes the converted state space graph, initial state, goal state and a
search strategy and return the corresponding solution/path according to the given strategy.
Please consider only breadth-first search and depth-first search strategies for this question.

```python
#Lets name our class "PathFinder"
class PathFinder:
    def __init__(self, ethiopian_cities, initial_state, goal_state, search_strategy):
        self.graph = ethiopian_cities
        self.initial_state = initial_state
        self.goal_state = goal_state
        self.search_strategy = search_strategy

    def solve(self):
        if self.search_strategy == "BFS":
            return self.bfs()
        elif self.search_strategy == "DFS":
            return self.dfs()
        else:
            raise ValueError("Invalid search strategy. Choose 'BFS' or 'DFS'.")

    def bfs(self):
        # Initialize a queue for BFS
        queue = [(self.initial_state, [self.initial_state])]  # Each element is (current_state, path_so_far)

        while queue:
            current_state, path_so_far = queue.pop(0)

            if current_state == self.goal_state:
                return path_so_far  # Found the goal state

            # Explore neighbors and add them to the queue
            for neighbor in self.graph[current_state]:
                if neighbor not in path_so_far:
                    queue.append((neighbor, path_so_far + [neighbor]))

        return "No solution found"  # Goal state not reachable

    def dfs(self):
        # Initialize a stack for DFS
        stack = [(self.initial_state, [self.initial_state])]  # Each element is (current_state, path_so_far)

        while stack:
            current_state, path_so_far = stack.pop()

            if current_state == self.goal_state:
                return path_so_far  # Found the goal state

            # Explore neighbors and add them to the stack
            for neighbor in self.graph[current_state]:
                if neighbor not in path_so_far:
                    stack.append((neighbor, path_so_far + [neighbor]))

        return "No solution found"  # Goal state not reachable





```

2.2 Assuming “Addis Ababa” as an initial state, write a program that use uniform cost search
to generate a path to “Lalibela”.

```python
import heapq

class PathFinderUsingUCS:
    def __init__(self, ethiopian_cities_with_cost, initial_state, goal_state):
        self.ethiopian_cities_with_cost = ethiopian_cities_with_cost
        self.initial_state = initial_state
        self.goal_state = goal_state

    def uniform_cost_search(self):
        # Initialize a priority queue (min-heap) for UCS
        pq = [(0, 0, self.initial_state, [self.initial_state])]  # Each element is (total_cost, path_cost, current_state, path_so_far)
        heapq.heapify(pq)

        while pq:
            total_cost, path_cost, current_state, path_so_far = heapq.heappop(pq)

            if current_state == self.goal_state:
                return path_so_far, path_cost  # Found the goal state

            # Explore neighbors and add them to the priority queue
            for neighbor, edge_cost in self.ethiopian_cities_with_cost[current_state].items():
                if neighbor not in path_so_far:
                    new_path_cost = path_cost + edge_cost
                    new_total_cost = total_cost + new_path_cost
                    heapq.heappush(pq, (new_total_cost, new_path_cost, neighbor, path_so_far + [neighbor]))

        return "No solution found", 0  # Goal state not reachable




initial_city = "Addis Ababa"
goal_city = "Lalibela"

solver = PathFinderUsingUCS(ethiopian_cities_with_cost, initial_city, goal_city)
solution_path, path_cost = solver.uniform_cost_search()
print(f"Solution path using UCS: {solution_path}")
print(f"Path cost: {path_cost}")
```

“Jimma”, “Bale”, “Sof Oumer”, and “Arba Minch” as goal states;in no specific order, write
a customized uniform cost search algorithm to generate a path that let a visitor visit all those
goal states preserving the local optimum.

```python
import heapq

class PathFinderUsingUCS:
    def __init__(self, ethiopian_cities_with_cost, initial_state, goal_states):
        self.ethiopian_cities_with_cost = ethiopian_cities_with_cost
        self.initial_state = initial_state
        self.goal_states = goal_states

    def uniform_cost_search(self):
        # Initialize a priority queue (min-heap) for UCS
        pq = [(0, self.initial_state, [self.initial_state])]  # Each element is (cost, current_state, path_so_far)
        heapq.heapify(pq)

        while pq:
            cost, current_state, path_so_far = heapq.heappop(pq)

            if current_state in self.goal_states:
                # If the current state is a goal state, return the path taken so far and its cost
                return path_so_far, cost

            # Explore neighbors and add them to the priority queue
            for neighbor, edge_cost in self.ethiopian_cities_with_cost[current_state].items():
                if neighbor not in path_so_far:
                    new_cost = cost + edge_cost
                    heapq.heappush(pq, (new_cost, neighbor, path_so_far + [neighbor]))

        return "No solution found", None  # Goal states not reachable



initial_city = "Addis Ababa"
goal_cities = {"Axum", "Gondar", "Lalibela", "Babile", "Jimma", "Bale", "Sof Oumer", "Arba Minch"}

solver = PathFinderUsingUCS(ethiopian_cities_with_cost, initial_city, goal_cities)
solution_path, path_cost = solver.uniform_cost_search()
print(f"Solution path using UCS: {solution_path}, Path cost: {path_cost}")
```

```python
# Define the locations and utilities
locations_of_Coffee = {
    "Addis Ababa": {"Ambo","Buta Jirra", "Adama"},
    "Ambo": {"Gedo", "Nekemte"},
    "Buta Jirra": {"Worabe", "Wolkite"},
    "Adama": {"Dire Dawa", "Mojo"},
    "Gedo": {"Shambu","Fincha"},
    "Nekemte": {"Gimbi", "Limu"},
    "Worabe": {"Hosana","Durame"},
    "Wolkite": {"Benchi Naji","Tepi"},
    "Mojo": {"Dilla","Kaffa"},
    "Dire Dawa": {"Chiro","Harar"}
}

utilities = {
    "Shambu": 4,
    "Fincha": 5,
    "Gimbi": 8,
    "Limu": 8,
    "Hosana": 6,
    "Durame": 5,
    "Benchi Naji": 5,
    "Tepi": 6,
    "Kaffa": 7,
    "Dilla": 9,
    "Chiro": 6,
    "Harar": 10
}
```

```python
class MiniMaxSearch:
    def __init__(self, locations_of_Coffee, utilities):
        self.locations_of_Coffee = locations_of_Coffee
        self.utilities = utilities

    def minimax(self, state, is_max, cost=0, path=[]):
        if state in self.utilities:
            return self.utilities[state] + cost, path + [state]

        if is_max:
            max_value = float('-inf')
            best_path = None
            for next_state in self.locations_of_Coffee[state]:
                value, new_path = self.minimax(next_state, not is_max, cost + 1, path + [state])
                if value > max_value:
                    max_value = value
                    best_path = new_path
            return max_value, best_path
        else:
            min_value = float('inf')
            best_path = None
            for next_state in self.locations_of_Coffee[state]:
                value, new_path = self.minimax(next_state, not is_max, cost + 1, path + [state])
                if value < min_value:
                    min_value = value
                    best_path = new_path
            return min_value, best_path

# Create an instance of MiniMaxSearch
searcher = MiniMaxSearch(locations_of_Coffee=locations_of_Coffee, utilities=utilities)

# Use MiniMax search to find the best path from "Addis Ababa"
value, path = searcher.minimax("Addis Ababa", True)

# Exclude the initial state from the path
path = path[1:] if path else path

# Print the best path and its cost
print("Best path:", path, "with cost:", value)
```
